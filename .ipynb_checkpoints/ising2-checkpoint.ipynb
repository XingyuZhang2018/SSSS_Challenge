```python
import numpy as np
import random
import matplotlib.pyplot as plt
import xlwt
import time

def init_spin_array(rows, cols):
    return np.ones((rows, cols))


def find_neighbors(array, lattice, x, y):
    left   = (x, y - 1)
    right  = (x, (y + 1) % lattice)
    top    = (x - 1, y)
    bottom = ((x + 1) % lattice, y)
    return [array[left[0], left[1]],
            array[right[0], right[1]],
            array[top[0], top[1]],
            array[bottom[0], bottom[1]]]


def energy(array, lattice, x ,y):
    return 2 * array[x, y] * sum(find_neighbors(array, lattice, x, y))

def IsingMC(temperature):
    spin_array = init_spin_array(lattice, lattice)
    # the Monte Carlo follows below
    mag = np.zeros(sweeps + RELAX_SWEEPS)
    for sweep in range(sweeps + RELAX_SWEEPS):
        for i in range(lattice):
            for j in range(lattice):
                e = energy(spin_array, lattice, i, j)
                if e <= 0:
                    spin_array[i, j] *= -1
                elif np.exp((-1.0 * e)/temperature) > random.random():
                    spin_array[i, j] *= -1
        mag[sweep] = abs(sum(sum(spin_array))) / (lattice ** 2)
          
    ma = sum(mag[RELAX_SWEEPS:]) / sweeps
    print(temperature, ma)
    x.append(temperature)
    y.append(ma)
    
lattice = 10 #eval(input("Enter lattice size: "))
RELAX_SWEEPS = 10000 #eval(input("Enter the number of Monte Carlo RELAX: "))
sweeps = 1000 #eval(input("Enter the number of Monte Carlo Sweeps: "))
TB = 1.0 #eval(input("Enter the begin of T: "))
TE = 4.0 #eval(input("Enter the end of T: "))
interval = 0.1 #eval(input("Enter the interval of T: "))
x =[]
y =[]
timex=time.time()
for temperature in np.arange(TB, TE,interval):
    IsingMC(temperature)
plt.figure(1)
plt.plot(x,y)
plt.show()
print (time.time()-timex)
```
