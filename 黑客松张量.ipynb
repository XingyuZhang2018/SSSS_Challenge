```python
import math
import numpy as np
```

```python
beta = 1
A = [[(math.sqrt(2*math.cosh(beta))-math.sqrt(2*math.sinh(beta)))/2,(math.sqrt(2*math.cosh(beta))+math.sqrt(2*math.sinh(beta)))/2],\
     [(math.sqrt(2*math.cosh(beta))+math.sqrt(2*math.sinh(beta)))/2,(math.sqrt(2*math.cosh(beta))-math.sqrt(2*math.sinh(beta)))/2]]
print(A)
```

```python
B = np.zeros(shape=(2,2,2))
for i in range(2):
    for j in range(2):
        for k in range(2):
            for n in range(2):
                B[i][j][k] += A[n][i]*A[n][j]*A[n][k]
print(B)
```

```python
C = np.zeros(shape=(2,2,2,2,2))
for a in range(2):
    for b in range(2):
        for c in range(2):
            for d in range(2):
                for e in range(2):
                    for i in range(2):
                        for j in range(2):
                            for k in range(2):
                                for l in range(2):
                                    for m in range(2):
                                        C[a][b][c][d][e] += B[a][i][j]*B[b][j][k]*B[c][k][l]*B[d][l][m]*B[e][m][i]
print(C)
```

```python
D = np.zeros(shape=(2,2,2,2,2,2,2,2,2,2))
for a in range(2):
    for b in range(2):
        for c in range(2):
            for d in range(2):
                for e in range(2):
                    for f in range(2):
                        for g in range(2):
                            for h in range(2):
                                for i in range(2):
                                    for j in range(2):
                                        for i0 in range(2):
                                            for i1 in range(2):
                                                for i2 in range(2):
                                                    for i3 in range(2):
                                                        for i4 in range(2):
                                                            for i5 in range(2):
                                                                for i6 in range(2):
                                                                    for i7 in range(2):
                                                                        for i8 in range(2):
                                                                            for i9 in range(2):
                                                                                    D[a][b][c][d][e][f][g][h][i][j] += C[i0][i1][i2][i3][i4]*C[i5][i0][i6][j][a]*C[i6][i1][i7][b][c]*C[i7][i2][i8][d][e]\
                                                                                    *C[i8][i3][i9][f][g]*C[i9][i4][i5][h][i]
print(D)
```

```python
Z = 0
for a in range(2):
    for b in range(2):
        for c in range(2):
            for d in range(2):
                for e in range(2):
                    for f in range(2):
                        for g in range(2):
                            for h in range(2):
                                for i in range(2):
                                    for j in range(2):
                                        Z += D[a][b][c][d][e][f][g][h][i][j] * D[b][c][d][e][f][g][h][i][j][a]
print(Z)
```
